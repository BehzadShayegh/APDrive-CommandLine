```
ZIP_NAME = 'zillow-prize-1.zip'
```

```
import pandas as pd
from zipfile import ZipFile

zip_file = ZipFile(DATA_PATH+ZIP_NAME)

path_properties_2016 = 'properties_2016.csv'
path_properties_2017 = 'properties_2017.csv'

# properties_2016 = pd.read_csv(zip_file.open(path_properties_2016))
properties_2017 = pd.read_csv(zip_file.open(path_properties_2017))
```

با خط کد زیر می توانیم به دیتای هر پارسل با آیدی آن دسترسی پیدا کنیم :

```
properties_2017.set_index('parcelid', inplace=True)
```

```
units_number, cols_number = properties_2017.shape
print(units_number, cols_number)
```

برای عملیات رگرسیون دیتای غیر عددی تاحدی مشکل ساز خواهد بود. پس آن ها را از دیتای مورد بررسی حذف می کنیم.

```
float_dtypes = (properties_2017.dtypes == 'float64')
float_dtypes
```

```
properties = (properties_2017[properties_2017.columns[float_dtypes.values]])
properties
```

همانطور که در بالا مشاهده می کنید دیتاست موجود دارای نقص های فراوانی است و نمی توان عملیات رگرسیون را به خوبی روی آنها انجام داد. پس در این قسمت با روشی ساده سعی می کنیم نقصان داده ها را با عدد ثابت -۱ پر کنیم :

```
import numpy as np
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='constant', fill_value=-1)

copleteProperties = imputer.fit_transform(properties.values)
copleteProperties
```

باید بتوانیم با آیدی پارسل ها به دیتای آن ها دسترسی داشته باشیم :

```
dictProperties = dict(zip(np.array(properties.index), copleteProperties))
dictProperties
```

بسیار خب. حالا زمان لود دیتای آموزش است. از دیتای سال ۲۰۱۶ به عنوان داده ی آموزش و از دیتای سال ۲۰۱۷ به عنوان داده ی تست استفاده خواهیم کرد. دلیل این امر آن است که ما درنهایت باید دیتای سال ۹۷ را پیشبینی کنیم :

```
import pandas as pd
from zipfile import ZipFile

zip_file = ZipFile(DATA_PATH+ZIP_NAME)

path_train_2016 = 'train_2016_v2.csv'
path_train_2017 = 'train_2017.csv'

train_2016 = pd.read_csv(zip_file.open(path_train_2016))
train_2017 = pd.read_csv(zip_file.open(path_train_2017))
```

```
import numpy as np

parcelid_2016 = np.array(train_2016['parcelid'])
parcelid_2017 = np.array(train_2017['parcelid'])

logerror_2016 = np.array(train_2016['logerror'])
logerror_2017 = np.array(train_2017['logerror'])
```

دیتای پارسل ها را با ترتیب دیتای آموزش و تست مرتب می کنیم :

```
from operator import itemgetter

orderX2016 = np.array([dictProperties[pid] for pid in parcelid_2016])
orderX2017 = np.array([dictProperties[pid] for pid in parcelid_2017])
```

```
(orderX2016.shape,
 logerror_2016.shape,
 orderX2017.shape,
 logerror_2017.shape)
```

حال زمان اجرای الگوریتم جنگل تصادفی می باشد :

```
from sklearn.ensemble import RandomForestRegressor

regr = RandomForestRegressor(n_estimators=1000)

regr.fit(orderX2016, logerror_2016)
print(regr.score(orderX2016, logerror_2016))
print(regr.score(orderX2017, logerror_2017))
```
